```python
import pandas as pd
from ctgan import CTGAN


# Load the 9-visit-per-mother synthetic timeline
df = pd.read_csv("Synthetic_Longitudinal_ANC.csv")

# Optional: drop Risk_Level if it's your RL agent's output
training_cols = df.drop(columns=["Risk_Level"])  # Or keep it if conditioning on risk

# Initialize CTGAN
ctgan = CTGAN(epochs=300)  # Increase epochs for better quality

# Fit the model
ctgan.fit(training_cols)

# Save the model for future reuse
ctgan.save("ctgan_model.pkl")
```

```python
# Generate 1000 new synthetic visit records (any mother, any visit)
samples = ctgan.sample(1000)

# Save or preview
samples.to_csv("CTGAN_Synthetic_ANC_Visits.csv", index=False)
samples.head()
```
